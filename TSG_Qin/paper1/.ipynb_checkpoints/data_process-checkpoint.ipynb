```python
import pandas as pd
tem = pd.read_csv('./Data/temperature.csv')
tem.head()
```

```python
tem = tem.rename(columns={"20191231":"day","1905":"minute","13.7":"temperature"})
tem['time']=tem['day'].astype(str)+tem['minute'].astype(str)
tem.drop(columns={'day','minute'},inplace=True)
tem.set_index('time', inplace = True)
```

```python
tem = tem[tem.index.str.len()==12]
tem = tem[tem['temperature'].str.len()<9]
tem = tem[tem['temperature'].astype(float)>-100]
```

```python
tem.index = pd.to_datetime(tem.index)
```
