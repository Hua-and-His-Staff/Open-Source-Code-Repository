```python
import torch
import numpy as np
import time
import matplotlib
#matplotlib.use('Agg')
import matplotlib.pyplot as plt

from Model import GridMLP
from VecEnv import EIEnv
from Train import MultivariateNormalDiag

env=EIEnv()
net = GridMLP(env.observation_space.shape[0],env.action_space.shape[0])
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
net = net.to(device)
```

```python
checkpoint = torch.load('model')
net.load_state_dict(checkpoint['model_state_dict'],strict=False)
net.eval()
```

```python
total_steps=2000
interval=20
init_cache=0.4154
seed=1926081
gamma=0.93


steps=[]
with torch.no_grad():
    observation = env.preset(init_cache,seed)
    for step in range(total_steps):
        observation = torch.from_numpy(observation).to(torch.float32).to(device)

        # network forward pass
        policies_mu, policies_sigma, value_net = net(observation)
        policies_prob = MultivariateNormalDiag(policies_mu,policies_sigma)
        #action = policies_prob.sample()
        action=policies_mu
        # gather env data, reset done envs and update their obs
        observation, reward, done, _ = env.step(action.cpu().numpy())
        steps.append((np.array([reward]), action.cpu().numpy(), observation, value_net.cpu().numpy()))


    final_observation = torch.from_numpy(observation).to(torch.float32).to(device)

    # network forward pass
    _, _, final_value_net = net(final_observation)
    steps.append((None, None, None,final_value_net))

n_step=len(steps)-1
V_x=final_value_net.data.cpu().numpy()
out=[None] * n_step
for i in reversed(range(n_step)):
    reward, action, observation,value_net=steps[i]
    V_x=reward+V_x*gamma
    out[i]=observation, env.action_map(action), reward, V_x, value_net

x,a,r,V_single,V_net=map(lambda l:np.concatenate(l),zip(*out))
x=x.reshape((-1,6))
a=a.reshape((-1,3))


steps=[]
observation = env.preset(init_cache,seed)
for step in range(total_steps):
    
    action=np.random.uniform(-1,1,3)
    # gather env data, reset done envs and update their obs
    observation, reward, done, _ = env.step(action)
    steps.append((np.array([reward]), env.action_map(action), observation))
                 
r_,a_,x_=map(lambda l:np.concatenate(l),zip(*steps))
x_=x_.reshape((-1,6))
a_=a_.reshape((-1,3))
```

```python
fig, (ax1)=plt.subplots(1,1,figsize=(12,6))
ax1.plot(V_single[::interval],'g',label='Simulation (V*)')
ax1.plot(V_net[::interval],'b--',label='Net prediction (V*)')
ax1.legend()
fig.tight_layout()
fig.show()
```
