```python
import matplotlib.pyplot as plt
import numpy as np
DGparas = np.array([[5e-1 ,5e-3, 200], [1, 3e-3, 300],[1.5, 2e-3, 400]])
for i in range(3):
    num = 10*int(DGparas[i,2])
    x = np.linspace(0, int(DGparas[i,2]), num=num, endpoint=False)
    y = np.zeros(num)
    for j in range(num):
        y[j] = DGparas[i,1]*x[j]**2 + DGparas[i,0]*x[j]
    plt.plot(x,y, label = i)
plt.xlim(0,400)
plt.ylim(0,1000)
plt.legend()
plt.grid()
plt.show()
```
